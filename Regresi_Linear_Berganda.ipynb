```python
import pandas as pd
import statsmodels.api as sm
```

```python
data = pd.read_excel("Data_Regresi_LinierBerganda.xlsx")

print(data.head())
```

```python
x = data[['X1','X2','X3','X4','X5','X6']]
x = sm.add_constant(x)
y = data['Y']
```

```python
model = sm.OLS(y, x).fit()

print_model = model.summary()
print(print_model)
```
